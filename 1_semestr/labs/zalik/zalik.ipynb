Створити клас ParityChecker, який:

- Приймає в конструкторі число
- Має метод check_parity(), який визначає           парність числа
- Має property parity, яка повертає рядок           "парне" або "непарне"


```python
class ParityChecker:
    def __init__(self, value):
        self.value = value

    def check_parity(self):
        return self.value % 2 == 0
    @property
    def parity(self):
        return "парне" if self.check_parity() else "непарне"
    
    def __str__ (self):
        return "Виконав студент групи КН-33 Середній Юрій"
if __name__ == "__main__":
    nums = [10, 15, 22, 33, 42]
    for n in nums:
        num = ParityChecker(n)
        print(f"Число {num.value} є {num.parity}.")
        print(num)
    
```
